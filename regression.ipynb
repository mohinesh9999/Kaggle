```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#gradient descent

X=np.arange(10)

Y=(X-5)**2
print(X,Y)
```

```python
#given a func f(x) we want to find the value of x that minimize f
```

```python
##visualization
```

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.ylabel("F(x)")
plt.xlabel("X")
```

```python
x=0
lr=0.1#step size
#take 50 steps in downhill direction
error=[]
for i in range(50):
  grad=2*(x-5)
  x=x-lr*grad
  e=(x-5)**2
  error.append(e)
  plt.plot(X,Y)
  plt.scatter(x,e)
  print(x)
  #break condition for threshold
```

```python
#plot value of error
plt.plot(error)
```
